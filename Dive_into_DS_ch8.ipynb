## Parsing HTML Code

### Scraping an Email Address

```python
#Scraping a webpage to find email addresses

urltoget = "https://bradfordtuckfield.com/contactscrape.html"
pagecode = requests.get(urltoget)


mail_beginning = pagecode.text.find("Email:")
print(mail_beginning)

#mail_end = pagecode.text.find(".com")
#print(mail_end)

#print(pagecode.text[mail_beginning:mail_end+4])
```

```python
#Trying to get the actual email

print(pagecode.text[mail_beginning:mail_beginning+80])
```

```python
#printing the exact email address

print(pagecode.text[mail_beginning+38:mail_beginning+64])
```

### Searching for Addresses Directly

```python
#searching for the @ symbol on the webpage

at_beginning = pagecode.text.find("@")
print(at_beginning)
```

```python
#printing the chars before and after the @ symbol

print(pagecode.text[at_beginning-4:at_beginning+22])
```

### Performing Searches with Regular Expressions

```python
#using regular expressions to find a email address in a webpage

import re

urltoget = "https://bradfordtuckfield.com/contactscrape2.html"
pagecode = requests.get(urltoget)

all_matches = re.finditer(r'[\w\.-]+@[\w\.-]+', pagecode.text)
print(all_matches)
```

```python
alladresses = []

for x in all_matches:
    alladresses.append(x[0] )

print(alladresses)
```

### Using Beautiful Soup

```python
#using beautiful soup to find all the hyperlinks in a page without regex

import requests
from bs4 import BeautifulSoup

URL = "https://bradfordtuckfield.com/indexarchive20210903.html"

response = requests.get(URL)
soup = BeautifulSoup(response.text, 'lxml')

for link in soup.find_all('a'):
    print(link.get('href'))
```
